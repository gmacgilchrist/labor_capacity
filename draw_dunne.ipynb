```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size']=14
```

```python
x = np.arange(24,34,0.1)
l = np.zeros(len(x))
xr = (x-25)
for i,xs in enumerate(x):
    xr = np.max([0,xs-25])
    l[i] = 100-25*xr**(2/3)
```

```python
fig,ax = plt.subplots(figsize=(5,5))
ax.plot(x,l,color='k',linestyle='--')
ax.set_xlim([24,34])
ax.set_ylim([0,100])
ax.set_ylabel('Labor Threshold Limit Value, $\lambda_d$ [%]')
ax.set_xlabel('WBGT$_d$ [$^\circ C$]')
fig.tight_layout()
fig.savefig('laborcapacity_functionalform.png',transparent=True,dpi=300)
```
